# ARIMA Model

ARIMA is an acronym that stands for **AutoRegressive Integrated Moving Average**. ARIMA model has the ability to make prediction based upon historical observation. It is popular and widely used statistical method for time series forecasting.

```python
emissionArima = emissionMain.copy()
dropColsArima = ['Coal, Including Coal Coke Net Imports, CO2 Emissions','Natural Gas, Excluding Supplemental Gaseous Fuels, CO2 Emissions','Distillate Fuel Oil, Excluding Biodiesel, CO2 Emissions','Motor Gasoline, Excluding Ethanol, CO2 Emissions','Residual Fuel Oil CO2 Emissions','Petroleum, Excluding Biofuels, CO2 Emissions','Other Carbon Emission']
emissionArima = emissionArima.drop(dropColsArima, axis=1)
emissionArima
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pylab import rcParams
%matplotlib inline

# Log to compress Scalability
trainLog = np.log(trainArima)
plt.figure(figsize=(20,7))
plt.plot(trainLog)
```

### Check Stationarity

**Stationarity** :- A stationary of time series is one whose statistical properties such as mean is all constant over time.

```python
# Rolling window over time period of One year, that is 12 months.

rolling_mean_trainArima = trainLog.rolling(window=12).mean()
rolling_mean_trainArima.head(13)
```

```python
original = plt.plot(trainLog, color='blue', label='Original')
mean = plt.plot(rolling_mean_trainArima, color='red', label='Rolling mean')

plt.legend(loc='best')
plt.title("Rolling Mean")
plt.show(block=False)
```

Here rolling mean over time is not constant, To test stationarity of the data we have,

*Null Hypothesis : Data is not stationary.*

*Atlternate Hypothesis : Data is stationary.*

For this, we have to perform Dicky-Fuller test.

```python
from statsmodels.tsa.stattools import adfuller

print('Results of Dickey-Fuller Test')
dftest = adfuller(trainLog['Total Energy CO2 Emissions'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#lags used', 'no. of observations used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value(%s)' %key] = value
    
print(dfoutput) 
```

Here p-value is 0.44 that is we more then accaptable p-value that is less then 0.05. So here we have to **accept the null hypothesis** that data is stationary. So to remove stationarity, we have method of differencing.

```python
trainArimaShift = trainLog - trainLog.shift()
plt.plot(trainArimaShift)
```

```python
from statsmodels.tsa.stattools import adfuller

#Function to check stationarity of emission data
def emission_stationarity(timeseries):
    #determining rolling ststistics
    movingAverage=timeseries.rolling(window=12).mean()
    
    #Plot rolling statistics
    orig = plt.plot(timeseries, color= 'blue', label = 'Original')
    mean = plt.plot(movingAverage, color = 'red', label = 'Rolling mean')
    plt.legend(loc='best')
    plt.title('Rolling mean')
    plt.show(block=False)
    
    #Dickey Fuller Test
    print('Results of Dickey Fuller Test')
    dftest = adfuller(timeseries["Total Energy CO2 Emissions"], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#lags used', 'no. of observations used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value(%s)' %key] = value
    print(dfoutput)
```

```python
trainArimaShift.dropna(inplace=True)
emission_stationarity(trainArimaShift)
```

After differencing of 1 position, the data is now stationary, as p-value is less than 0.05. So here we can accept the null hypothesis.

```python
#ACF and PACF plots
from statsmodels.tsa.stattools import acf,pacf

lag_acf = acf(trainArimaShift, nlags=40)
lag_pacf = pacf(trainArimaShift, nlags=40, method='ols')
#ols is Ordinary Least Squared method
#Plot ACF:
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.stem(lag_acf)
plt.axhline(y=0, linestyle='--', color='grey')
plt.axhline(y=-1.96/np.sqrt(len(trainArimaShift)), linestyle='--', color='gray') # Critical Value of ACF
plt.axhline(y=1.96/np.sqrt(len(trainArimaShift)), linestyle='--', color='gray') # Critical Value of ACF
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.stem(lag_pacf)
plt.axhline(y=0, linestyle='--', color='grey')
plt.axhline(y=-1.96/np.sqrt(len(trainArimaShift)), linestyle='--', color='gray') # Critical Value of PACF
plt.axhline(y=1.96/np.sqrt(len(trainArimaShift)), linestyle='--', color='gray') # Critical Value of PACF
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

First, we fit an ARIMA(2,1,2) model. This sets the lag value to 2 for autoregression, uses a difference order of 1 to make the time series stationary, and uses a moving average model of 2.

```python
from statsmodels.tsa.arima_model import ARIMA

#AR Model
model = ARIMA(trainLog, order=(2,1,2))
#(p,d,q)
#d = number of shift
results_ARIMA = model.fit()
plt.plot(trainArimaShift)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f' % sum((results_ARIMA.fittedvalues-trainArimaShift['Total Energy CO2 Emissions'])**2))
#RSS stands for residual sum of square
print('Plotting AR Model')
```

For better model we have to decrease the RSS value.

```python
from statsmodels.tsa.arima_model import ARIMA

#AR Model
model = ARIMA(trainLog, order=(11,1,6))
#(p,d,q)
results_ARIMA = model.fit()
plt.plot(trainArimaShift)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f' % sum((results_ARIMA.fittedvalues-trainArimaShift['Total Energy CO2 Emissions'])**2))
# RSS = residual sum of square
print('Plotting AR Model')
```

So for the RSS value drops to 0.50. So the trained model is better then the previous one.

# SARIMA Model

```python
emissionSarima = emissionMain.copy()
dropColsArima = ['Coal, Including Coal Coke Net Imports, CO2 Emissions','Natural Gas, Excluding Supplemental Gaseous Fuels, CO2 Emissions','Distillate Fuel Oil, Excluding Biodiesel, CO2 Emissions','Motor Gasoline, Excluding Ethanol, CO2 Emissions','Residual Fuel Oil CO2 Emissions','Petroleum, Excluding Biofuels, CO2 Emissions','Other Carbon Emission']
emissionSarima = emissionSarima.drop(dropColsArima, axis=1)
emissionSarima
```

```python
percent_training = 0.85
splitPoint = round(len(emissionSarima) * percent_training)
training, testing = emissionSarima[0:splitPoint], emissionSarima[splitPoint:]
```

Here Split data in Training and Testing Data and took log to decrease scalibility. After that we have done differencing by 1 variable.

```python
training = np.log(training)
training_diff = training.diff(periods=1).values[1:]

plt.plot(training_diff)
```

```python
#ACF and PACF plots
from statsmodels.tsa.stattools import acf,pacf

lag_acf = acf(training_diff, nlags=40)
lag_pacf = pacf(training_diff, nlags=40, method='ols')
#ols is Ordinary Least Squared method
#Plot ACF:
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.stem(lag_acf)
# plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='grey')
plt.axhline(y=-1.96/np.sqrt(len(training_diff)), linestyle='--', color='gray') # Critical Value for ACF
plt.axhline(y=1.96/np.sqrt(len(training_diff)), linestyle='--', color='gray') # Critcal Value for PACF
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.stem(lag_pacf)
plt.axhline(y=0, linestyle='--', color='grey')
plt.axhline(y=-1.96/np.sqrt(len(training_diff)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(training_diff)), linestyle='--', color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

Also from ACF graph here we can say that on every 12th lag the trend is repeating. So We have Seasonal Pamater S=12. Since this lag is Positive so that P=1 and Q=0.

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(training, order=(11,1,6), seasonal_order=(1,1,0,12), enforce_stationarity=False, enfoce_invertibility=False)
model_fit = model.fit(disp=False)
```

```python
K  = len(testing)
forecast = model_fit.forecast(K)
forecast = np.exp(forecast)

plt.figure(figsize=(20,10))
plt.plot(forecast,'r')
plt.plot(emissionSarima,'b')
plt.autoscale(enable=True, axis='x', tight=True)
plt.axvline(x=emissionSarima.index[splitPoint], color='black');
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(testing, forecast)
print("rmse = ",sqrt(MSE))
```

In ARIMA we have encountered the RMSE value 158 and For SARIMA it drops to 16. So the trained model is better then the previous one.

```python
# Now we can extend our prediction to next 24 months and se how trend is going on with current graph.
predict = model_fit.predict(36,600)
exp_predict = np.exp(predict)
exp_predict

plt.figure(figsize=(20,10))
plt.plot(emissionSarima)
plt.plot(exp_predict)
```
